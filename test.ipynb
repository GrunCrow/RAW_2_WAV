```python
import os
import wave
```

```python
# read .raw files and convert to .wav files with specified parameters


raw_path = "Archivo/raws/20251003-UCACAZALLA-000001.raw"
wav_path = "Archivo/raws_wavs/wavs/20251003-UCACAZALLA-000001.wav"

def convert_raw_to_wav(raw_file, wav_file, n_channels=1, sampwidth=2, framerate=44100):
    with open(raw_file, 'rb') as rf:
        raw_data = rf.read()
    
    with wave.open(wav_file, 'wb') as wf:
        wf.setnchannels(n_channels)
        wf.setsampwidth(sampwidth)
        wf.setframerate(framerate)
        wf.writeframes(raw_data)
convert_raw_to_wav(raw_path, wav_path)
print(f"Converted {raw_path} to {wav_path}")
```

```python
# read wav file metadata to verify parameters values and print
wav_path = "Archivo/wavs/20251004-UCACAZALLA-000021.wav"

with wave.open(wav_path, 'rb') as wf:
    n_channels = wf.getnchannels()
    sampwidth = wf.getsampwidth()
    framerate = wf.getframerate()
    n_frames = wf.getnframes()
    duration = n_frames / framerate
    print(f"Channels: {n_channels}, Sample Width: {sampwidth}, Frame Rate: {framerate}, Frames: {n_frames}, Duration: {duration:.2f} seconds")
```
